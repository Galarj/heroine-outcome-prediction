```python
import matplotlib.pyplot as plt
import pandas as pd 
stories = pd.read_excel('../data/stories_cleaned.xlsx')
heroines = pd.read_excel('../data/candidates_cleaned.xlsx')
df = pd.merge(heroines, stories, on="story_id")
```

## Part 1: Story-Level Features
These features describe the world and structure of the story itself. They set the rules of the romance game!

importance for each column


**medium**: The medium dictates how much time the story has to develop characters. Light Novels are notorious for long-running harem setups, while Visual Novels offer branching routes (where everyone can win, or no one does!).

**demographic**: Demographics carry heavy narrative expectations. Shounen romances often lean into high-comedy harem tropes, Seinen can be more experimental or mature, while Shoujo romances almost always focus on a single, clear male lead with high emotional drama.



Which Stories Have the Largest Romantic Casts?

```python
# story with the most candidates

most_candidates = df.groupby("title")["candidate_name"].count()
top10 = most_candidates.nlargest(10)

plt.figure(figsize=(10,6))
plt.barh(top10.index, top10.values)

plt.title("Top 10 most heroines")
plt.xlabel("number of heroines")
plt.ylabel("Title")
```

What Source Material Does the Dataset Cover?

```python
manga_count= (stories["medium"]== "Manga").sum()
anime_count = (stories["medium"]== "Anime").sum()
light_novel_count = (stories["medium"] == "Light Novel").sum()
visual_novel_count = (stories["medium"] == "Visual Novel").sum()
print(manga_count)
print(anime_count)
print(light_novel_count)
print(visual_novel_count)


# wtf this is a command ?? op as shit
medium_counts = stories["medium"].value_counts()
print(medium_counts)


plt.figure(figsize=(10,6))
plt.title("Medium Counts")
plt.bar(medium_counts.index, medium_counts.values)
plt.xlabel("Medium")
plt.ylabel("Number counts")
```

Which Demographics Are Most Common?

```python
demographic_count= stories["demographic"].value_counts()
print(demographic_count)
plt.figure(figsize=(10,6))

plt.bar(demographic_count.index, demographic_count.values)

plt.title("All Demographics and how many are there")
plt.xlabel("demographic")
plt.ylabel("How many")
```

What Types of Romance Are Represented?

```python
stories["romance_type"].unique()
romance_type_count = stories["romance_type"].value_counts()

plt.figure(figsize=(10,6))
plt.barh(romance_type_count.index, romance_type_count.values)
plt.xlabel("how many")
plt.ylabel("Genre")

plt.title("Romance Type Distribution")
```

What Hair Colors Are Most Common Among Heroine Candidates?

```python
hair_color_values = df["hair_color"].value_counts()
plt.figure(figsize=(10,6))
plt.barh(hair_color_values.index, hair_color_values.values)
plt.xlabel("number of hair color")
plt.ylabel("hair color ")
plt.title("Hair Color Distribution")
```

```python
archetype_count = df["primary_archetype"].value_counts()
print(archetype_count)
plt.figure(figsize=(10,6))
plt.barh(archetype_count.index, archetype_count.values)
plt.title("archetype")
```

```python
candidate_length = df.groupby("title")["candidate_name"].count()
plt.figure(figsize=(8,5))

plt.hist(candidate_length, bins="auto")

plt.title("Distribution of Heroine Counts per Story")
plt.xlabel("Number of Heroine Candidates")
plt.ylabel("Number of Stories")

plt.show()

```
