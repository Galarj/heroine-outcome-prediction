```python
import matplotlib.pyplot as plt
import pandas as pd 
stories = pd.read_excel('../data/stories_cleaned.xlsx')
heroines = pd.read_excel('../data/candidates_cleaned.xlsx')
df = pd.merge(heroines, stories, on="story_id")
```

```python
ml_df = df.drop(columns=[
    "candidate_id",
    "candidate_name",
    "confidence_score",
    "reasoning",
    "commitment_status",
    "first_chapter_or_episode",
    "confidence_score"

])
```

```python
stories["heroine_count"].mean()
#Heroine count greater than the dataset mean (4.43).
#Stories with more heroines than the dataset average may distribute romantic focus 
# across more characters, reducing the probability of any single heroine winning
```

High Heroine Count

```python
mean_heroine = stories["heroine_count"].mean()
ml_df["high_heroine_count"] = (
    ml_df["heroine_count"] > mean_heroine
).astype(int)
```

```python
ml_df["high_heroine_count"]
```

introduced early

```python
ml_df["introduced_early"] = (
    ml_df["introduction_order"] <= 3
).astype(int)
```

childhood friend

```python
ml_df["has_childhood_connection"] = (
    ml_df["childhood_connection"].isin(["Childhood Friend", 
        "Childhood Promise",
        "Acquaintance"
    ])
).astype(int)
```

```python
# if modern series
ml_df["is_modern"] = (
    ml_df["release_year"] >= 2010
).astype(int)
```
